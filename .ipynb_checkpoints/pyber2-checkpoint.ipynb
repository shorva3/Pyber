```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

city = pd.read_csv("city_data.csv")
ride = pd.read_csv("ride_data.csv")
```

```python
c_df = pd.merge(city, ride, how = 'outer', on = 'city')
c_df.head()
```

```python
urban = c_df.loc[c_df['type'] == 'Urban']
suburban = c_df.loc[c_df['type'] == 'Suburban']
rural = c_df.loc[c_df['type'] == 'Rural']
```

```python
avg_urban_fare = urban.groupby(['city'])['fare'].mean()
urban_total_rides = urban.groupby(['city'])['ride_id'].count()
urban_total_drivers = urban.groupby(['city'])['driver_count'].value_counts()

avg_suburban_fare = suburban.groupby(['city'])['fare'].mean()
suburban_total_rides = suburban.groupby(['city'])['ride_id'].count()
suburban_total_drivers = suburban.groupby(['city'])['driver_count'].value_counts()

avg_rural_fare = rural.groupby(['city'])['fare'].mean()
rural_total_rides = rural.groupby(['city'])['ride_id'].count()
rural_total_drivers = rural.groupby(['city'])['driver_count'].value_counts()
```

```python
plt.scatter(urban_total_rides,avg_urban_fare, s = urban_total_drivers*10, color = 'red', edgecolor = 'black', alpha = 0.5, label = 'Urban')
plt.scatter(suburban_total_rides,avg_suburban_fare, s = suburban_total_drivers*10, color = 'blue', edgecolor = 'black', alpha = 0.5, label = 'Suburban')
plt.scatter(rural_total_rides, avg_rural_fare, s = rural_total_drivers*10, color = 'yellow', edgecolor = 'black', alpha = 0.5, label = 'Rural')
plt.grid()
plt.legend(loc = 'best')
```

```python
total_fares = c_df.groupby(['type'])['fare'].sum()
colors = ['yellow', 'blue', 'red']
explode = [0,0,0.1]
labels = ['Rural', 'Suburban', 'Urban']
plt.pie(total_fares, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True)
```

```python
total_rides = c_df.groupby(['type'])['ride_id'].count()
plt.pie(total_rides, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True)
```
